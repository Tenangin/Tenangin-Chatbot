```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('evaluasi-tenobot.csv')

# Hilangkan kolom non-numerik dan kolom identitas untuk fokus pada evaluasi kuantitatif
numerical_df = df.select_dtypes(include=[np.number])

# Statistik deskriptif
stats = numerical_df.describe().T[['mean', 'std', 'min', 'max']]

# Visualisasi rata-rata tiap pertanyaan
plt.figure(figsize=(10, 8))
mean_scores = numerical_df.mean().sort_values()
sns.barplot(x=mean_scores.values, y=mean_scores.index, palette="viridis")
plt.title("Rata-rata Penilaian Tiap Aspek Evaluasi Tenobot")
plt.xlabel("Rata-rata Skor (1-5)")
plt.ylabel("Pertanyaan Evaluasi")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

stats.sort_values("mean", ascending=False)
```
